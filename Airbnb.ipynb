**Task 4: Data Analysis**

Univariate Analysis

```python
# Identifying outliers in price column using boxplot
df2 = df[df["price"] <1600]
sns.boxplot(data=df2, y="price")
```

```python
# Rows where price is greater than 1600
df[df["price"] >1600]
```

```python
# Number of rows where price is greater than 1600
df[df["price"] >1600].shape
```

```python
# Price distribution
plt.figure(figsize=(20,5))
sns.histplot(data=df2, x="price", bins=150)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```

```python
# Identifying outliers in availability_365 column using boxplot
sns.boxplot(data=df2, x="availability_365")
```

```python
# Availability_365 distribution
plt.figure(figsize=(20,5))
sns.histplot(data=df2, x="availability_365", bins=50)
plt.xlabel("Availability")
plt.ylabel("Frequency")
plt.title("Availability_365")
plt.show()
```

```python
# Average price for each neighbourhood_group
df2.groupby(by= "neighbourhood_group")["price"].mean()
```

```python
# Average price for each neighbourhood
df2.groupby(by= "neighbourhood")["price"].mean()
```

**Task 5: Feature Engineering**

```python
# Creating a new column 'price_per_bed'
df2["price_per_bed"] = df2["price"]/df2["beds"]
```

```python
df2.head()
```

```python
# Average price_per_bed for each neighbourhood_group
df2.groupby(by= "neighbourhood_group")["price_per_bed"].mean()
```

```python
# Average price_per_bed for each neighbourhood
df2.groupby(by= "neighbourhood")["price_per_bed"].mean()
```

Bivariate Analysis

```python
df2.columns
```

```python
# Neighbourhood_group and Price Relationship
plt.figure(figsize=(20,5))
sns.barplot(data=df2, x="neighbourhood_group", y="price", hue="room_type")
plt.title("Locality Vs Price")
plt.xlabel("Locality")
plt.ylabel("Price")
plt.show()
```

```python
# Number of Reviews and Price Relationship
plt.figure(figsize=(20,5))
sns.scatterplot(data=df2, x="number_of_reviews", y="price", hue="neighbourhood_group")
plt.title("Number of Reviews Vs Price")
plt.xlabel("Number of Reviews")
plt.ylabel("Price")
plt.show()
```

```python
# Pair plots of different numerical columns
plt.figure(figsize=(20,20))
sns.pairplot(data=df2, vars=("price", "minimum_nights", "number_of_reviews", "availability_365",), hue="room_type")
plt.show()
```

```python
# Airbnb geographical listings
plt.figure(figsize=(20,10))
sns.scatterplot(data=df2, x="longitude", y="latitude", hue="room_type")
plt.title("Airbnb Geographical Listings")
plt.show()
```

```python
df2.dtypes
```

```python
# Correlations between differnet numeric columns
corr = df2[["latitude", "longitude", "price", "minimum_nights", "availability_365", "price_per_bed"]].corr()
corr
```

```python
# Heatmap for the correlation
plt.figure(figsize=(5,5))
sns.heatmap(data=corr, annot=True)
plt.title("Heatmap for correlation")
plt.show()
```
